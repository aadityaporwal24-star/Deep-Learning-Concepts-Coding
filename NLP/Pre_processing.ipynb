```python
#Stop Words Removal
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
stop= stopwords.words("english")
stop_words= stop + list(punctuation)
print(stop_words)
sent= "This is a sample sentence, showing off the stop words filtration." 
sent_new=word_tokenize(sent)          
for i in sent_new:
    if i not in stop_words:
        print(i)
```

```python
#Stemming
import nltk
from nltk.stem import PorterStemmer,RegexpStemmer, SnowballStemmer,LancasterStemmer
l=LancasterStemmer()
r=RegexpStemmer("ing")
s= SnowballStemmer("english")
p=PorterStemmer()
print(l.stem("changed"))
print(r.stem("changing"))
print(p.stem("changing"))
print(s.stem("changing"))
```

```python
#Lemmatization
from nltk.stem import WordNetLemmatizer
wl= WordNetLemmatizer()
print(wl.lemmatize("studies"))
```

```python
#n-grams
import nltk
x="I am Aaditya Porwal I am learning NLP I am a Tech Student"
from nltk.tokenize import word_tokenize
w=word_tokenize(x)
from nltk.collocations import ngrams

bigrams= ngrams(w,2)
trigrams= ngrams(w,3)
print(list(bigrams))
print(list(trigrams))
```

```python
#Count Vectorizer
l=["I am learning NLP","NLP is interesting","I am a student"]
import sklearn
import pandas as pd 
df=pd.DataFrame({"name":l})
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer()
new_data=cv.fit_transform(df["name"]).toarray()

print(cv.vocabulary_)
```

```python
#Word Sense Disambiguation 
x= "The sun rises in the east and sets in the west because the Earth rotates eastward"
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
w=word_tokenize(x)
print(lesk(w,"rise").definition())
```
