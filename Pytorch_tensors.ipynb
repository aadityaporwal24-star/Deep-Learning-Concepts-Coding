```python
import torch
x=torch.tensor([1.,2.,3.,4.])
print("x:",x)

y=torch.tensor([2.,4.,6.,8.])
print("y:",y)
w=torch.tensor(0.0)
b=torch.tensor(0.0)
print("w:",w)
print("b:",b)   

y_pred=w*x +b
print("y_pred:",y_pred)
loss=((y_pred-y)**2).mean()
print("loss:",loss)
```
